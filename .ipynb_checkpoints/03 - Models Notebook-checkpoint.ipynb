## Load the data
The collected data is broken down into the 2014-2015, 2016-2017, 2017-2018, and current 2018-2019 seasons. The first four full seasons will be loaded in and concatenated into one large training set (83,415 rows).  The current season will serve as the test set (10,497 rows).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read in each season, drop the Unnamed: 0 column and make the date the index

df_15 = pd.read_csv('./data/five_game14-15.csv')
df_15.rename(columns={'Unnamed: 0': 'date'}, inplace=True)
df_15.index = df_15.date
df_15.drop(columns='date', axis=1, inplace=True)

df_16 = pd.read_csv('./data/five_game15-16.csv')
df_16.rename(columns={'Unnamed: 0': 'date'}, inplace=True)
df_16.index = df_16.date
df_16.drop(columns='date', axis=1, inplace=True)

df_17 = pd.read_csv('./data/five_game16-17.csv')
df_17.rename(columns={'Unnamed: 0': 'date'}, inplace=True)
df_17.index = df_17.date
df_17.drop(columns='date', axis=1, inplace=True)

df_18 = pd.read_csv('./data/five_game17-18.csv')
df_18.rename(columns={'Unnamed: 0': 'date'}, inplace=True)
df_18.index = df_18.date
df_18.drop(columns='date', axis=1, inplace=True)

df_19 = pd.read_csv('./data/five_game18-19.csv')
df_19.rename(columns={'Unnamed: 0': 'date'}, inplace=True)
df_19.index = df_19.date
df_19.drop(columns='date', axis=1, inplace=True)
```

```python
#make train and test sets
train = pd.concat([df_15, df_16, df_17,
                  df_18], axis = 0)

test = df_19

print(train.shape)
print(test.shape)
```

## Regression Models

I tried Linear Regression, Random Forest, AdaBoost, GradientBoost, and Keras Regressor models, with $R^2$ scores on the test data ranging from .518 to .551.  Utlimately here I will just show output from The Linear Regression, Random Forest, and Gradient Boosted models. 

The only model that was markedly overfit was the Random Forest.  The rest of the models had $R^2$ scores that differed by no more than .02 between train, and test data.

Root mean squared error, the metric by which I ultimately evaluated each model stayed in the 9.6 - 10 range.

```python
#these featues do not need to be scaled
#gs and home are binary and the other two
#are just averages of the target variable 
#from past observations

no_scale = ['gs', 'home', 'shifted_expand', 'fantasy_roll5']

#features to be scaled.  many are rolling five game
#averages or advanced stats regarding the opponent

to_scale = ['mp_roll5', 'fga_roll5', 'fta_roll5', 'trb_roll5',
            'stl_roll5', 'blk_roll5',
            'tov_roll5', 'pf_roll5', '3p_pct_roll5',
            'ft_pct_roll5', 'fg_pct_roll5', '3pa_roll5',
            'ast_roll5', 'dub_dub_roll5',
            'usg_pct_roll5', 'plus_minus_roll5',
            'ortg_roll5', 'opp_avg_drtg', 'opp_avg_pace',
            'opp_avg_o_tov_pct']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#setting the X and y variables. X_train and X_test
#will start with the features that do not need to be
#scaled

ss = StandardScaler()

X_train = train[no_scale]
X_test = test[no_scale]
y_train = train.fantasy_points
y_test = test.fantasy_points

#then we scale the rest and add them back to the
#train and test sets

scaled_train = pd.DataFrame(ss.fit_transform(train[to_scale]),
                     columns = train[to_scale].columns,
                     index = train[to_scale].index)

X_train = pd.concat([X_train, scaled_train], axis = 1)

scaled_test = pd.DataFrame(ss.fit_transform(test[to_scale]),
                     columns = test[to_scale].columns,
                     index = test[to_scale].index)

X_test = pd.concat([X_test, scaled_test], axis = 1)
```

```python
#visualizing the target variable in the train data

plt.figure(figsize=(10, 8))
plt.title('NBA 2014-2018 Seasons (Training Data)', fontsize=30)
plt.xlabel('Fantasy Points', fontsize=25)
plt.ylabel('Games',fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

#add the median fantasy score
plt.axvline(y_train.mean(), color='r', linestyle='dashed', linewidth=1)


plt.hist(y_train, alpha=0.5, histtype='bar', ec='black')
plt.show();
```

```python
#visualizing target variable in test data.  very similar 
#distribution

plt.figure(figsize=(10, 8))
plt.title('NBA 2018-1019 Season (Test Data)', fontsize=30)
plt.xlabel('Fantasy Points', fontsize=25)
plt.ylabel('Games',fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.axvline(y_test.mean(), color='r', linestyle='dashed', linewidth=1)
plt.hist(y_test, alpha=0.5, histtype='bar', ec='black')
plt.show();
```

```python
from sklearn.metrics import mean_squared_error
```

```python
#establish baseline Root Mean Squared Error
test['mean_y'] = np.mean(y_test)
rmse = mean_squared_error(y_test, test['mean_y']) ** .5
rmse
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

# Random Forest

I didn't do a full on grid search, but I did try fitting models with 100, 200, 300, and 400 estimators, and 300 worked the best out of those four options.

```python
rf = RandomForestRegressor(n_estimators=300)
rf_model = rf.fit(X_train, y_train)
```

```python
rf_train = rf_model.score(X_train, y_train).round(3)
rf_test = rf_model.score(X_test, y_test).round(3)
print('The R2 score is ' + str(rf_train) \
     + ' on the train data and ' + str(rf_test) + ' on the test data')
```

```python
#get predictions on train and test data to visualize the overfitting below
rf_train_preds = rf_model.predict(X_train)
rf_test_preds = rf_model.predict(X_test)

#get the root mean squared error on test predictions
rf_rmse = (mean_squared_error(y_test, rf_test_preds) ** .5).round(3)
print('The root mean squared error on test data is ' \
     + str(rf_rmse))
```

```python
#visualize the overfit 
plt.figure(figsize=(20, 14))

#scatter plot of true y value vs predicted
plt.scatter(y_train, rf_train_preds, c='b', s=8, alpha = .5)

#plot the line going through the true y values
plt.plot(y_train, y_train, c='r')

plt.title('Random Forest \n Predictions on the Training Data',
         fontsize=35)
plt.ylabel('Predicted Score', fontsize=30)
plt.xlabel('Actual Score', fontsize=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30);
```

```python
plt.figure(figsize=(20, 14))

#scatter plot of true y value vs predicted
plt.scatter(y_test, rf_test_preds, c='b', s=8, alpha = .5)

#plot the line going through the true y values
plt.plot(y_test, y_test, c='r')

plt.title('Random Forest \n Predictions on the Test Data',
         fontsize=35)
plt.ylabel('Predicted Score', fontsize=30)
plt.xlabel('Actual Score', fontsize=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30);
```

# Gradient Boost

In this case I tried the same range of estimators as I did with Random Forest and 200 was the best result.  You can see the $R^2$ scores on the training and test data are just about dead on with one another.

```python
grad = GradientBoostingRegressor(n_estimators=200)
grad_model = grad.fit(X_train, y_train)
```

```python
grad_train = grad_model.score(X_train, y_train).round(3)
grad_test = grad_model.score(X_test, y_test).round(3)
print('The R2 score is ' + str(grad_train) \
     + ' on the train data and ' + str(grad_test) + ' on the test data')
```

```python
grad_train_preds = grad_model.predict(X_train)
grad_test_preds = grad_model.predict(X_test)

grad_rmse = (mean_squared_error(y_test, grad_test_preds) ** .5).round(3)
print('The root mean squared error on test data is ' \
     + str(grad_rmse))
```

```python
from sklearn.linear_model import LinearRegression
```

# Linear Regression

Using the features from the previous two regressor models didn't work quite as well here, so I used different features that had low p-values according to stats models and they performed better in the linear regression model.  Conversely the features that performed better in the linear regression did not do so well in the random forest or gradient boost models.

```python
#using a slightly different set of features on linear regression models
#these features did not score as high on the above models, but do better
#with linear regression

linear_to_scale = ['gs_roll5', 'mp_roll5',
            'fg_roll5', 'fga_roll5', 'fg_pct_roll5',
            '3pa_roll5', 'ft_roll5', 'fta_roll5',
            'orb_roll5', 'drb_roll5', 'trb_roll5',
            'ast_roll5', 'stl_roll5', 'blk_roll5',
            'tov_roll5', 'pts_roll5', 'plus_minus_roll5',
            'efg_pct_roll5', 'ast_pct_roll5', 
            'trip_dub_roll5', 'dub_dub_roll5',
            'opp_score_roll5', 'opp_ortg_roll5',
            'opp_pace_roll5', 'opp_trb_pct_roll5'
]
```

```python
X_train = train[no_scale]
X_test = test[no_scale]
y_train = train.fantasy_points
y_test = test.fantasy_points
```

```python
scaled_train = pd.DataFrame(ss.fit_transform(train[linear_to_scale]),
                     columns = train[linear_to_scale].columns,
                     index = train[linear_to_scale].index)

X_train = pd.concat([X_train, scaled_train], axis = 1)

scaled_test = pd.DataFrame(ss.fit_transform(test[linear_to_scale]),
                     columns = test[linear_to_scale].columns,
                     index = test[linear_to_scale].index)

X_test = pd.concat([X_test, scaled_test], axis = 1)
```

```python
lr = LinearRegression()
lr_model = lr.fit(X_train, y_train)
```

```python
lr_train = lr_model.score(X_train, y_train).round(3)
lr_test = lr_model.score(X_test, y_test).round(3)
print('The R2 score is ' + str(lr_train) \
     + ' on the train data and ' + str(lr_test) + ' on the test data')
```

```python
lr_train_preds = lr_model.predict(X_train)
lr_test_preds = lr_model.predict(X_test)

lr_rmse = (mean_squared_error(y_test, lr_test_preds) ** .5).round(3)
print('The root mean squared error on test data is ' \
     + str(lr_rmse))
```

```python
#output dataframes of all three models' predictions and errors
#generate an aggregate of all three models to see if it predicts
#any better

to_average = ['random_forest', 'gradient_boosted', 'linear_regression']

predictions = pd.DataFrame(columns=['random_forest',
                                    'gradient_boosted',
                                    'linear_regression',
                                    'combined', 'actual_scores'],
                          index = y_test.index)

predictions['random_forest'] = rf_test_preds.round(2)
predictions['gradient_boosted'] = grad_test_preds.round(2)
predictions['linear_regression'] = lr_test_preds.round(2)
predictions['combined'] = predictions[to_average].mean(1).round(2)
predictions['actual_scores'] = y_test

predictions.head()
```

```python
combined_rmse = mean_squared_error(y_test, predictions.combined) ** .5

models = ['random_forest',
          'gradient_boosted',
          'linear_regression',
          'combined']

all_rmse = [rf_rmse, grad_rmse, lr_rmse, combined_rmse]

errors = pd.DataFrame(columns=['model', 'rmse'])

errors['model'] = models
errors['rmse'] = all_rmse

errors
```

The lowest root mean squared error (albeit by just .016 over linear regression) came from taking the mean predictions from each of the three models.

## Classification

To see if classification models do any better, I'll split the types of performace into 4 tiers.  The first tier will be those that get 50 or more fantasy points in a game, the second, between 30 and 50, the third between 20 and 30, and the fourth under 20.  These are slightly arbitrary, but I am going a bit off of the summary statistics of the target variable.  

```python
#the spread of the target variable differs a bit in the
#train and test sets, but not much

display(train.fantasy_points.describe())
display(test.fantasy_points.describe())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
```

```python
#first assign the labels to the data
#choose arbitrarily large numbers for the top of the 
#tier 1 range and bottom of the tier 4 range

criteria = [train['fantasy_points'].between(50, 1000), 
            train['fantasy_points'].between(30, 50),
            train['fantasy_points'].between(20, 30),
            train['fantasy_points'].between(-1000, 20)]
values = [1, 2, 3, 4]

train['tier'] = np.select(criteria, values, 0)
```

```python
test['tier'] = test['fantasy_points']

criteria = [test['fantasy_points'].between(50, 1000), 
            test['fantasy_points'].between(30, 50),
            test['fantasy_points'].between(20, 30),
            test['fantasy_points'].between(-1000, 20)]
values = [1, 2, 3, 4]

test['tier'] = np.select(criteria, values, 0)
```

```python
#check the baseline accuracy for the models
#if the model picked 4 every time it would be
#have an accuracy score of .48
test.tier.value_counts(normalize=True)
```

```python
y_train = train.tier
y_test = test.tier
```

```python
rf = RandomForestClassifier(n_estimators=300)
rf_class = rf.fit(X_train, y_train)
```

```python
rf_class_train = rf_class.score(X_train, y_train).round(3)
rf_class_test = rf_class.score(X_test, y_test).round(3)
print('The accuracy score is ' + str(rf_class_train) \
     + ' on training data and ' + str(rf_class_test) \
     + ' on test data')
```

Just like the random forest regressor, this model overfits to the training data in a big way.

```python
from sklearn.metrics import confusion_matrix
```

```python
rf_class_preds = rf_class.predict(X_test)


rf_conmat = confusion_matrix(y_test, rf_class_preds)

# put confusion matrix in a dataframe
rf_conmat = pd.DataFrame(rf_conmat,
                      columns=['Predicted '+ str(cls) for cls in rf_class.classes_],\
                      index=['Actual ' + str(cls) for cls in rf_class.classes_])
rf_conmat
```

```python
grad = GradientBoostingClassifier(n_estimators=200)
grad_class = grad.fit(X_train, y_train)
```

```python
grad_class_train = grad_class.score(X_train, y_train).round(3)
grad_class_test = grad_class.score(X_test, y_test).round(3)
print('The accuracy score is ' + str(grad_class_train) \
     + ' on training data and ' + str(grad_class_test) \
     + ' on test data')
```

```python
grad_class_preds = grad_class.predict(X_test)


grad_conmat = confusion_matrix(y_test, grad_class_preds)

grad_conmat = pd.DataFrame(grad_conmat,
                      columns=['Predicted '+ str(cls) for cls in grad_class.classes_],\
                      index=['Actual ' + str(cls) for cls in grad_class.classes_])
grad_conmat
```
